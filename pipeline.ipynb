```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline

```

```python
df=pd.read_csv("C:\\Users\\Dell\\Downloads\\datasets\\Disease_Dataset.csv")
df.head()
```

```python
columns_to_drop=['has_night_blindness','has_fatigue','has_bleeding_gums','has_bone_pain','has_muscle_weakness','has_numbness_tingling','has_memory_problems','has_pale_skin','has_multiple_deficiencies',"diet_type","sun_exposure","income_level","latitude_region","folate_percent_rda","calcium_percent_rda","iron_percent_rda","hemoglobin_g_dl","serum_vitamin_d_ng_ml","serum_vitamin_b12_pg_ml","serum_folate_ng_ml","symptoms_count",'vitamin_a_percent_rda','vitamin_c_percent_rda','vitamin_d_percent_rda','vitamin_e_percent_rda','vitamin_b12_percent_rda','symptoms_list']
df=df.drop(columns=columns_to_drop)
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=["disease_diagnosis"]),df["disease_diagnosis"],test_size=0.2,random_state=42)
```

```python
ohe_cols = ['gender']
oe=['smoking_status',
    'alcohol_consumption',
    'exercise_level']
smoking_status=['Never','Former','Current']
alcohol_consumption=['Moderate','Heavy']
exercise_level=['Sedentary','Light','Moderate','Active']
```

Applying Column Transformation

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', Pipeline([
            ('ohe', OneHotEncoder(
                drop='first',
                handle_unknown='ignore'
            ))
        ]), ohe_cols),

        ('ord',Pipeline([
            ('imputer',SimpleImputer(strategy='most_frequent')),
            ('ordinal',OrdinalEncoder(
                categories=[
                    'smoking_status',
                    'alcohol_consumption',
                    'exercise_level'
                ]
            ))
        ]),oe)
        
    ],
    remainder='drop'  
)
```

Forming Pipeline

```python
pipe = Pipeline([
    ('preprocessor', preprocessor)
])
```

Applying Pipeline on train test dataset

```python
pipe.fit(x_train, y_train)
```
