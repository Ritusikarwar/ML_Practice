```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\Dell\\Downloads\\datasets\\placement.csv")
df
```

```python
# Only for Simple Linear Regression
df=df.drop(columns=['placed'])
```

```python
df
```

Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(df[['cgpa']],df[['placement_exam_marks']],test_size=0.2,random_state=42)
```

```python
#Applying LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.plot(x_test,lr.predict(x_test),color='red')
plt.xlabel('CGPA')
plt.ylabel('Placement_exam_marks')
```
