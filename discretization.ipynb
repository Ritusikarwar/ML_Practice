```python
df.head()
```

```python
columns_to_drop=["date","waterfront","view","condition","sqft_basement","sqft_above","yr_renovated","street","city","statezip","country"]
df=df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
columns_to_drop=["bathrooms","sqft_lot","floors"]
df=df.drop(columns=columns_to_drop)
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['price']),df["price"],test_size=0.2,random_state=42)
```

Applying Regression before binning

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

Accuracy

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
