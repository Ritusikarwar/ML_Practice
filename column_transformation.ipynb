```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("C:\\Users\\Dell\\Downloads\\datasets\\Disease_Dataset.csv")
df.head()
```

Extracting meaningful columns

```python
columns_to_drop=['has_night_blindness','has_fatigue','has_bleeding_gums','has_bone_pain','has_muscle_weakness','has_numbness_tingling','has_memory_problems','has_pale_skin','has_multiple_deficiencies',"diet_type","sun_exposure","income_level","latitude_region","folate_percent_rda","calcium_percent_rda","iron_percent_rda","hemoglobin_g_dl","serum_vitamin_d_ng_ml","serum_vitamin_b12_pg_ml","serum_folate_ng_ml","symptoms_count",'vitamin_a_percent_rda','vitamin_c_percent_rda','vitamin_d_percent_rda','vitamin_e_percent_rda','vitamin_b12_percent_rda','symptoms_list']
df=df.drop(columns=columns_to_drop)
df.head()
```

```python
column_to_drop=["alcohol_consumption"]
df=df.drop(columns=column_to_drop)
df.head()
```

Train test splitting

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=["disease_diagnosis"]),df["disease_diagnosis"],test_size=0.2,random_state=42)
```

column Transformation

```python
transformer=ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(sparse_output=False,drop='first'),['gender']),
    ('tnf2',OrdinalEncoder(categories=[
        ["Never","Former","Current"],
        ["Sedentary","Light","Moderate","Active"]]),['smoking_status','exercise_level']),
],remainder='passthrough')
#remainder can be passthrough(ignores the rest columns) or drop() drops the rest columns 
```

```python
x_train=transformer.fit_transform(x_train)
x_test=transformer.transform(x_test)
```

```python
le = LabelEncoder()
y_train= le.fit_transform(y_train)
y_test= le.transform(y_test)
```
