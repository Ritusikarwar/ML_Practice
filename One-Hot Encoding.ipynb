```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("C:\\Users\\Dell\\Downloads\\datasets\\customer.csv")
df.head()
```

Train Test splitting

```python
x=df[['gender']] #dataframe (2D)
y=df['purchased'] #dataframe (1D)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
type(x_train)
```

Applying One Hot Encoding on input column

```python
ohe=OneHotEncoder()
x_train_new=ohe.fit_transform(x_train[["gender"]]).toarray()
x_train_new
```

```python
x_test_new=ohe.transform(x_test[["gender"]]).toarray()
x_test_new
```

Applying Label Encoding on output column

```python
le=LabelEncoder()
le.fit(y_train)
y_train_new=le.transform(y_train)
y_train_new
```

```python
y_test_new=le.transform(y_test)
y_test_new
```
