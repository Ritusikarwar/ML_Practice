```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("C:\\Users\\Dell\\Downloads\\datasets\\linear_regression.csv")
df.head()
df.info()
```

```python
df=df.dropna()
```

```python
x=df[['x']]
y=df['y']
plt.scatter(x,y)
plt.show()
```

By OLS

```python
lr=LinearRegression()
lr.fit(x,y)
lr.coef_ #m
```

```python
lr.intercept_ 
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red')
```

By Gradient Descent
Assumptions:-
1) Let m be constant throughout that is m=1.00065638
2) Let, b=0 end goal b=-0.107265

```python
lr.intercept_  
# we would try to find intercept using gradient descent
```

```python
y_pred=(1.00065638*x)+1
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label="OLS")
plt.plot(x,y_pred,color='blue',label='b=1')
plt.legend()
plt.show()
```

```python
m=1.00065638
b=1
loss_slope=-2*np.sum(y-m*x-b)
loss_slope
```

Incoperating learning rate

```python
learning_rate=0.1
step_size=loss_slope*learning_rate
step_size
```

Calculating the new intercept

```python
b=b-step_size
b
```

```python
y_pred2=(1.00065638*x)+b
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='pink',label='b={}'.format(b))
plt.plot(x,y_pred2,color='blue',label='b=1')
plt.legend()
plt.show()
```
