```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("C:\\Users\\Dell\\Downloads\\datasets\\linear_regression.csv")
df.head()
df.info()
```

```python
df=df.dropna()
```

```python
x=df[['x']]
y=df['y']
plt.scatter(x,y)
plt.show()
```

By OLS

```python
lr=LinearRegression()
lr.fit(x,y)
lr.coef_ #m
```

```python
lr.intercept_ 
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red')
```
